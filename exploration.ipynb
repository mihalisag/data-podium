## Colors

```python
import fastf1
import fastf1.plotting

schedule = get_schedule_until_now()
events = list(schedule['EventName'])
```

```python
events
```

```python
session = fastf1.get_session(2024, 'Sao Paulo Grand Prix', 'R')
session.load()
```

```python
session.laps['Driver'].unique()
```

```python
session.laps['Team'].unique()
```

```python
fastf1.plotting.get_driver_color('VER', session=session)
```

```python
teams = session.laps['Team'].unique()

TEAM_COLORS = {team: fastf1.plotting.get_team_color(team, session) for team in teams}
```

```python
TEAM_COLORS.update({'ass':1})
```

```python
TEAM_COLORS
```

```python
fastf1.plotting.get_driver_color_mapping(session)
```

```python
DRIVER_COLORS = {}
TEAM_COLORS = {}

for event in events:

    session = fastf1.get_session(2024, event, 'R')
    session.load()

    # Driver colors
    DRIVER_COLORS.update(fastf1.plotting.get_driver_color_mapping(session))
    
    # Team colors 
    teams = session.laps['Team'].unique()
    TEAM_COLORS.update({team: fastf1.plotting.get_team_color(team, session) for team in teams})
```

```python
DRIVER_COLORS
```
